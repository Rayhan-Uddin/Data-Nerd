# Logical Operators/ Condition

```python
x=5
y=6
z=11
print(x<y and x+y==z) #Returns True if both statements are true
```

```python
x=5
y=6
z=11
print(x>y or x+y==z)#Returns True if one of the statements is true
```

```python
x=5
y=6
z=11
print(not(x>y or x+y==z)) #Reverse the result, returns False if the result is true
```
